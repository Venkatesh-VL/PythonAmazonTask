```python
from selenium import webdriver
import pandas as pd
import time
import os
```

```python

driverPath = r"C:\Users\merit\Downloads\chrd\chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverPath)
driver.get("https://www.amazon.in/s?k=headphones+offer&crid=O6NXEPY5KTI&sprefix=headphones+off%2Ccomputers%2C355&ref=nb_sb_ss_ts-doa-p_1_14")
title = driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
original_price = driver.find_elements_by_xpath("//span[@class='a-price a-text-price']")
offer_price = driver.find_elements_by_xpath("//span[@class='a-price']")
lst = []
lst1 = []
lst2 = []
writer = pd.ExcelWriter(r"C:\Users\merit\Documents\PythonAssignments\Headphones.xlsx", engine='xlsxwriter')
for i in title:
    lst.append(i.text)
print(lst)
for j in original_price:
    lst1.append(j.text)
print(lst1)
for k in offer_price:
    lst2.append(k.text)
print(lst2)


df = pd.DataFrame(list(zip(lst,lst1,lst2)), columns = ['Product Name','Original Price','Offer Price'])
df.to_excel(writer,index = False)
writer.save()
writer.close()
print(df)
```
